```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Step(x):
    return np.array(x>0, dtype = np.int)  # 조건에 따라 참/거짓, 1/0의 값을 반환
```

```python
def Sigmoid(X):
    return 1 / (1 + np.exp(-X))
```

```python
x = np.arange(-10.0,10.0,0.1)
y1 = Step(x)
y2 = Sigmoid(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid()
plt.show()
```
