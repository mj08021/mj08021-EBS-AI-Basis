```python
# 데이터 준비하기
import pandas as pd
```

```python
df = pd.read_csv('temp_ice.csv', encoding = 'euc-kr')
```

```python
# 데이터를 입력 변수와 출력 변수로 나누기
import numpy as np
```

```python
data = np.array(df)
```

```python
X = data[:,1]
```

```python
Y = data[:, -1]
```

```python
# 입력 변수와 출력 변수 각각의 평균(mean) 구하기
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
# X변수의 개수 구하기
n = len(X)
```

```python
# 최소제곱법을 이용하여 beta0과 beta1 구하기
temp1 = 0
temp2 = 0
```

```python
for i in range(n):
    temp1 += (X[i] - mean_x) * (Y[i] - mean_y)
    temp2 += (X[i] - mean_x) ** 2
```

```python
beta1 = temp1 / temp2
beta0 = mean_y - (beta1 * mean_x)
```

```python
# 학습 결과 시각화하기
import matplotlib.pyplot as plt
Y_pred = beta0 + beta1 * X

plt.title('Avg_temp & Clicks')
plt.xlabel('Average temperature(C)')
plt.ylabel('Clicks')
plt.plot(X, Y, 'k.')
plt.plot(X, Y_pred, color = 'red')
plt.axis([-4, 30, 20, 100])
plt.grid()
plt.show()
```
