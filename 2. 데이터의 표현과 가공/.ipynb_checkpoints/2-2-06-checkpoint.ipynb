```python
# 외부 모듈 정의
import numpy as np  # 벡터, 행렬 데이터를 쉽게 처리하기 위한 모듈 선언
import matplotlib.pyplot as plt  # 화면에 이미지 데이터를 보여주기 위한 모듈 선언
import PIL.Image as pilimg
```

```python
im = pilimg.open('sea.bmp')  # image file 읽어오기
```

```python
pix = np.array(im)  # image dataf를 numpy array로 구성하기
```

```python
pixSize = np.array(pix.shape)
```

```python
print(pixSize)
```

```python
# pix array에서 각각 R(0), G(1), B(2) 성분값 외에는 0으로 만든 후
# 원본 이미지에서 R, G, B에 해당하는 배열 만들기
pix_R = pix.copy()
pix_R[:,:,(1,2)] = 0
pix_G = pix.copy()
pix_G[:,:,(0,2)] = 0
pix_B = pix.copy()
pix_B[:,:,(0,1)] = 0
```

```python
# 원본 이미지인 pix 행렬을 이미지 데이터로 출력
plt.subplot(411)
plt.imshow(pix)
plt.axis('off')
plt.title('RGB')
```

```python
# pix 행렬에서 이미지 데이터의 R채널 출력
plt.subplot(412)
plt.imshow(pix_R)
plt.axis('off')
plt.title('RGB')
```

```python
# pix 행렬에서 이미지 데이터의 G채널 출력
plt.subplot(413)
plt.imshow(pix_G)
plt.axis('off')
plt.title('G(Green)')
```

```python
# pix 행렬에서 이미지 데이터의 B채널 출력
plt.subplot(414)
plt.imshow(pix_B)
plt.axis('off')
plt.title('B(Blue)')
```

```python
plt.show()
```
