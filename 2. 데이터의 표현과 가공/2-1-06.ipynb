```python
import csv
```

```python
a = [[], [], [], [], [], [], []]
```

```python
with open('passby_data.CSV', 'r') as f:
    reader = csv.DictReader(f)
    i = j = 0
    for row in reader :
        a[i].append(row)
        j = j + 1
        if(j % 24 == 0):
            i = i + 1
```

```python
x_title = ['MON', 'TUE', 'WED', 'THR', 'FRI', 'SAT', 'SUN']
```

```python
for i in range(0, 7):
    for j in range(0, len(a[i])):
        print(x_title[i], '[', j, '] = ', a[i][j])
```
